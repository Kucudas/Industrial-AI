MOT(Multi-Object Tracking)
- 다중 객체 추적
- 비디오 시퀀스에서 다수의 객체를 실시간으로 추적하는 기술
- 각 객체가 프레임 사이에서 이동할 때 고유한 식별자(ID)를 유지
- 각 프레임에서의 위치를 지속적으로 추적
- 위치 예측, 객체 재식별(Re-ID), 데이터 연관(Data Association) 등의 기술을 사용

MOT 구성 요소
1. 객체 탐지(Object Detection)
 - 각 프레임에서 객체를 감지하고 이를 바운딩 박스나 마스크 등으로 표시
 - 탐지된 객체는 각 프레임마다 위치 정보로 활용

2. 위치 예측(State Estimation)
 - 칼만 필터와 같은 상태 추정 기법을 사용하여 객체의 위치와 속도를 고려한 다음 위치를 예측
 - 상수 속도 모델을 주로 사용

3. 데이터 연관(Data Association)
 - 각 프레임마다 새롭게 탐지된 객체와 이전 프레임의 객체 간에 최적의 매칭을 수행
 - 헝가리안 알고리즘, IoU(Intersection over Union), Re-ID 모델을 통한 외형 기반 매칭
 - 프레임 간 객체가 일치한다고 판단되면 동일한 ID를 유지하며, 동일한 객체임을 식별

4. 객체 재식별(Re-ID)
 - 밀집된 환경이나 구조물에 가려지는 상황에서 객체가 프레임 밖으로 사라졌다가 다시 나타나는 경우가 많음. 이 경우 이전에 추적하던 객체가 맞는지 확인하기 위해 Re-ID 모델을 사용
 - Re-ID는 객체의 외형을 분석하여, 동일 객체 여부를 판단 --> CNN 기반 특징 추출 모델을 사용하여 객체를 분류

5. 트랙 관리(Track Management)
 - 객체가 프레임에 나타났다가 사라질 때 이를 처리하기 위해 트랙 생성과 삭제 과정이 필요
 - 트랙 관리는 프레임에서 새로 나타난 객체를 추적 대상으로 추가하고, 일정 시간 동안 탐지되지 않은 객체는 추적에서 제외하는 방식으로 관리

 MOT 주요 기법
 1. Tracking-by-Detection (탐지 기반 추적)
  - 매 프레임마다 객체를 탐지하고, 이전 프레임의 객체와 매칭하여 추적을 수행
  - 객체 탐지기의 성능에 크게 의존하며, 고성능 탐지기와 결합할 때 더 높은 정확도를 보임

2. Online vs Batch Tracking
 - 온라인 추적(Online Tracking): 실시간으로 각 프레임을 처리하며, 과거 프레임 정보만을 이용하여 현재 프레임을 추적
 - 배치 추적(Batch Tracking): 여러 프레임을 한꺼번에 처리하며, 미래 정보까지 고려해 객체 간 매칭을 최적화

3. SORT(Simple Online and Realtime Tracking)
 - 단순하고 빠른 온라인 MOT 알고리즘으로, 칼만 필터와 헝가리안 알고리즘을 사용하여 탐지된 객체와 추적 객체 간의 매칭을 수행
 - SORT는 빠르고 간단하지만 외형 정보(appearance)를 사용하지 않으므로, 밀집된 환경에서 객체가 서로 겹칠 때 ID가 전환되는 문제점이 있음

4. DeepSORT
 - DeepSORT는 SORT에 Re-ID 기능을 추가하여 객체의 외형 정보를 학습하여 ID 스위치를 줄임
 - CNN 기반의 외형 특징 추출기를 이용해 객체의 ID를 지속적으로 유지할 수 있어 밀집 환경에서도 더 높은 성능을 나타냄

5. BoT-SORT
 - ORT의 확장판으로, 카메라 움직임 보정 기능을 추가하고, Re-ID 모델을 통합하여 더 정확한 데이터 연관을 수행
 - Kalman 필터의 상태 벡터를 개선하여 예측 정확도를 높였으며, 다중 객체 추적 성능을 더욱 향상

BoT-SORT(Baseline on Track-SORT): 기존의 SORT 모델을 기반으로, 다중 객체 추적(MOT)에서 더 높은 정확도와 안정성을 목표로 하는 개선된 모델. 객체의 외형 정보를 통한 Re-ID와 카메라 움직임 보정 기능을 추가하여 객체 재식별 성능을 높임

BoT-SORT 주요 특징
1. SORT(Simple Online and Realtime Tracking) 기반
 - SORT는 위치 정보만을 기반으로 하기 때문에 객체 간의 ID 전환(ID Switch)이 발생하기 쉬움
 - 특히 밀집된 환경이나 객체가 가려질 때, ID가 잘못 매칭되는 경우가 많음
 - SORT 모델을 기반으로, SORT 모델의 한계를 보완함

2. Re-ID(재식별) 기능 통합
 - 객체의 외형 정보(appearance)를 추출하여 Re-ID 기능을 강화
 - CNN 기반의 Re-ID 네트워크를 사용해 객체의 특징을 추출하고, 이러한 특징 정보를 바탕으로 추적 중인 객체와 탐지된 객체를 매칭하여, ID 전환을 줄이고 일관된 추적을 유지
 - IoU(Intersection over Union) 기반의 매칭에 Re-ID 정보를 결합하여 객체 간 유사도를 높여 정확한 매칭을 수행

3. 카메라 움직임 보정(Camera Motion Compensation, CMC)
 - 카메라가 움직이는 환경에서 발생하는 문제를 해결하기 위해 카메라 모션 보정 기능을 추가
 - 이미지 등록(Image Registration)과 같은 기법을 사용하여 프레임 간 카메라의 움직임을 보정함으로써, 객체의 상대적 위치를 보다 정확하게 예측
 - 이동하는 카메라나 불안정한 배경에서도 안정적인 추적이 가능함

4. 개선된 Kalman 필터 설정
 - 기존 SORT 모델의 칼만 필터는 객체의 중심 위치와 크기만을 예측하는 상수 속도 모델임
 - BoT-SORT는 이 필터의 상태 벡터를 수정하여 바운딩 박스의 너비와 높이를 직접 예측하도록 개선
 - 객체의 바운딩 박스를 더욱 정확하게 추적하도록 돕고, 예측의 정확도를 높임

BoT-SORT 구성 요소
1. IoU-ReID 융합 매칭
 - IoU와 Re-ID 유사도를 결합하여, 탐지된 객체와 추적 중인 객체 간의 매칭을 수행
 - 코사인 거리(cosine distance)를 사용하여 Re-ID 특징 간의 유사도를 계산하고, IoU와 융합
 - 밀집된 환경에서 객체 간 혼동을 줄이고, 가려졌다가 다시 나타나는 객체의 ID를 유지하는 데 유리함

2. 트랙 관리 (Track Management)
 - 특정 조건에 따라 객체의 추적 트랙을 생성하고 삭제
 - 탐지된 객체가 일정 시간 동안 추적되지 않으면 추적 대상에서 제외하고, 새로운 객체가 탐지되면 새로운 ID를 부여하여 트랙을 생성
 - Kalman 필터 예측 결과와 Re-ID 모델의 유사도를 결합하여 객체의 일관된 ID를 유지

4. Re-ID 모델
 - FastReID와 같은 Re-ID 모델을 사용하여 객체의 외형 정보를 학습
 - Re-ID 모델은 객체의 외형적 특징을 추출하여 각 객체를 고유하게 식별할 수 있도록 돕고, 객체가 가려졌다가 다시 나타날 때 동일 객체로 인식하는 데 중요한 역할을 함
